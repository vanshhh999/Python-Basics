Q.2 What is an interpreter in Python?
- In Python, an interpreter is a program that executes Python code directly, translating it into machine code (or an intermediate format) step by step, as the program runs. This contrasts with compilers, which translate the entire program into machine code before execution begins.

Q.3 What are pre-defined keywords in Python?
- In Python, predefined keywords (also known as reserved words) are words that have a special meaning in the language and cannot be used for variable names, function names, or identifiers. These keywords are reserved by the Python interpreter because they define the structure and behavior of the language.

Q.4 Can keywords be used as variable names?
- No, keywords in Python cannot be used as variable names or identifiers. Keywords are reserved words that have a special meaning in the Python programming language and are integral to the syntax and structure of the language.

Q.5 What is mutability in Python?
- Mutability in Python refers to the ability of an object to be changed after it is created. An object is said to be mutable if its contents (like the values inside) can be modified. On the other hand, an object is immutable if it cannot be changed once it is created.

Q.6 Why are lists mutable, but tuples are immutable?
- Lists (Mutable): Lists are designed to store collections of items where the content is likely to change. For example, you might want to add, remove, or modify elements in a list after it is created. Lists are built to be flexible and allow for modifications, making them ideal for situations where the collection of items is dynamic and needs to be updated frequently.
- Tuples (Immutable): Tuples, on the other hand, are designed for situations where the data should remain constant. The immutability of tuples guarantees that once they are created, their contents cannot be altered. This makes tuples useful for representing fixed collections of items that should not change, such as coordinates (x, y), dates, or function return values that should remain consistent.

Q.7 What is the difference between “==” and “is” operators in Python?
- 1. == (Equality Operator)
The == operator is used to check equality of values. It compares the contents or values of two objects to see if they are the same. This means that the == operator checks if the data stored inside two objects is equal, regardless of whether they are the same object in memory.

- 2. is (Identity Operator)The is operator is used to check identity. It compares whether two objects are the same object in memory, i.e., if they occupy the same memory location. If two variables point to the same object in memory, is returns True; otherwise, it returns False.

Q.8 What are logical operators in Python?
- In Python, logical operators are used to perform logical operations on Boolean values (i.e., True or False). These operators allow you to combine conditional statements or check multiple conditions at once.

Q.9 What is type casting in Python?
- Type casting in Python refers to the process of converting one data type into another. It allows you to convert a variable of one type (e.g., an integer) into a different type (e.g., a string or a float). This process is essential when working with different types of data and when you need to perform operations that require specific data types.

Q.10 What is the difference between implicit and explicit type casting?
- 1. Implicit Type Casting (Automatic Type Casting)
- Implicit type casting occurs automatically when Python converts one data type to another without the need for the programmer’s intervention. Python performs this type of conversion when it is safe and does not result in loss of data.
- 2. Explicit Type Casting (Manual Type Casting)
- Explicit type casting occurs when the programmer manually converts one data type to another using specific functions. This type of casting is required when you need to perform a conversion that Python does not automatically handle.

Q.11 What is the purpose of conditional statements in Python?
- Conditional statements in Python are used to perform different actions based on whether a certain condition is True or False. They allow the program to make decisions and control the flow of execution. This is fundamental in programming, as it enables the program to react differently depending on the inputs, environment, or conditions it encounters.

Q.12 How does the elif statement work?
- The elif statement in Python stands for "else if" and is used to check multiple conditions in sequence. It allows you to test several conditions and execute different blocks of code depending on which condition is True.

Q.13 What is the difference between for and while loops?
- 1. for loop:
- The for loop is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block of code a fixed number of times. It is often used when the number of iterations is known ahead of time.

- 2. while loop:
- The while loop is used to repeatedly execute a block of code as long as a specified condition is True. The condition is checked before each iteration, and the loop continues running until the condition becomes False.

Q.14 Write a program to reverse a string using a while loop.
- # Function to reverse a string using a while loop
def reverse_string(input_string):
    # Initialize variables
    reversed_string = ""
    index = len(input_string) - 1  # Start from the last character
    
    # Use a while loop to reverse the string
    while index >= 0:
        reversed_string += input_string[index]  # Add character to the reversed string
        index -= 1  # Move to the previous character
    
    return reversed_string

# Input string from the user
user_input = input("Enter a string: ")

# Call the function and print the reversed string
reversed_str = reverse_string(user_input)
print("Reversed string:", reversed_str)

Q.15 Write a Python program to calculate the factorial of a number provided by the user using a while loop.
- # Function to calculate the factorial of a number using a while loop
def factorial(n):
    result = 1  # Start with a result of 1
    while n > 1:  # Continue as long as n is greater than 1
        result *= n  # Multiply result by the current value of n
        n -= 1  # Decrease n by 1
    return result

# Get input from the user
user_input = int(input("Enter a number to calculate its factorial: "))

# Call the function and print the result
fact = factorial(user_input)
print(f"The factorial of {user_input} is: {fact}").
